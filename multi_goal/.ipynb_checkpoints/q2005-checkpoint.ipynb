```python
from gridworld import *
from tqdm import trange
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#2005 Q-learning


goals = [(0, 5), (4, 1)]
error_states = [(o, 0) for o in range(6)] + [(5, o) for o in range(1, 6)]
gw = gridworld((6, 6), goals, error_states, 0.79, 0, 1, 0)

goals = [(0, 7), (6,1)]
error_states = [(o, 0) for o in range(8)] + [(7,o) for o in range(1, 7)]
gw = gridworld((8, 8), goals, error_states, 0.79, 0, 1, 0)

# goals = [(0, 6), (5,1)]
# error_states = [(o, 0) for o in range(7)] + [(6,o) for o in range(1, 6)]
# gw = gridworld((7, 7), goals, error_states, 0.79, 0, 1, 0)

# goals = [(0, 9), (8,1)]
# error_states = [(o, 0) for o in range(10)] + [(9,o) for o in range(1, 9)]
# gw = gridworld((10, 10), goals, error_states, 0.79, 0, 1, 0)
```

```python
# #Empirically estimate risk probability
# num_sims = 10000
# def estimate_rr(pi):
#     risky = 0
#     rewards = []
#     for i in trange(num_sims):
#         s = gw.reset()
#         reward = 0
#         d = False
#         j = 0
#         while(j < 100):
#             a = pi[s[0],s[1]]
#             s1,r,d = gw.step(a)
#             #print(s)
#             reward += r
#             if d == True:
#                 if s1 in gw.error_states:
#                     risky += 1
#                 break
#             s = s1
#         rewards.append(reward)
#     return sum(rewards)/float(num_sims), float(risky)/num_sims    

def estimate_rr(pi):
    num_sims = 100
    all_rewards = []
    all_risks = []
    for st in [(2,3)]:#gw.enum_good_states: 
        #print(st)
        risky = 0
        rewards = []
        for i in trange(num_sims):
            s = st
            gw.set_state(s)
            reward = 0
            d = False
            j = 0
            print("new ep")
            while(j < 100):
                a = pi[s[0],s[1]]
                #print("from", s, "through", a)
                s1,r,d = gw.step(a)
                print("from", s, "to", s1, "through", a)
                reward += r
                if d == True:
                    if s1 in gw.error_states:
                        risky += 1
                        #print("o")
                    break
                s = s1
            rewards.append(reward)
        st_reward = sum(rewards)/float(num_sims)
        all_rewards.append(st_reward)
        st_risk = float(risky)/num_sims  
        all_risks.append(st_risk)
        #print(st_reward, st_risk)
    return sum(all_rewards)/len(all_rewards), sum(all_risks)/len(all_risks)
    
```

```python
trials = 3
Q_list = []
reward_trial_list = []

np.random.seed(1)

# Set learning parameters
lr = 0.9
y = 0.9
num_episodes = 10000
factor = 1
#create lists to contain total rewards and steps per episode

for k in trange(trials):
    #Initialize table with all zeros
    Q = np.zeros((gw.M, gw.N, 4))
    Q_ = np.zeros((gw.M, gw.N, 4))
    mainQ = np.zeros((gw.M, gw.N, 4))
    jList = []
    rList = []
    for i in range(num_episodes):
        
        if i%10000 == 1:
            print('Episode', i)            
        #Reset environment and get first new observation
        s = gw.reset()
        rAll = 0
        d = False
        j = 0
        #The Q-Table learning algorithm
        while j < 100:
            j+=1
            #Choose an action by greedily (with noise) picking from Q table
            a = np.argmax(Q[s[0],s[1],:] + np.random.randn(1,4)*(1./(i+1)))
#             if np.random.rand() < epsilon:
#                 # take random action
#                 a = np.random.choice([0,1,2,3])
#             else:
#                 # take action according to the q function table
#                 a = np.argmax(Q[s[0],s[1],:])
            #Get new state and reward from environment
            s1,r,d = gw.step(a)

            r_ = 0
            if s1 in gw.error_states:
                r_ = 1

            #Update Q-Table with new knowledge
            Q[s[0], s[1],a] = Q[s[0], s[1],a] + lr*(r + y*np.max(Q[s1[0], s1[1],:]) - Q[s[0], s[1],a])
            Q_[s[0], s[1],a] = Q_[s[0], s[1],a] + lr*(r_ + y*np.max(Q_[s1[0], s1[1],:]) - Q_[s[0], s[1],a])
            mainQ = factor * Q - Q_  

            rAll += r
            s = s1
            if d == True:
                break
        jList.append(j)
        rList.append(rAll)
        
    Q_list.append(mainQ)

#     v = np.mean(mainQ, axis = 2)
#     gw.print_v(v)


    pi = np.argmax(mainQ, axis = 2)
    gw.print_policy(pi)
    reward_trial_list.append(sum(rList)/float(num_episodes))
    #print "Score over time: " +  str(sum(rList)/float(num_episodes))
    
#     window = int(num_episodes/10)

#     plt.figure(figsize=[9,16])
#     plt.subplot(411)
#     plt.plot(pd.Series(jList).rolling(window).mean())
#     plt.title('Step Moving Average ({}-episode window)'.format(window))
#     plt.ylabel('Moves')
#     plt.xlabel('Episode')

#     plt.subplot(412)
#     plt.plot(pd.Series(rList).rolling(window).mean())
#     plt.title('Reward Moving Average ({}-episode window)'.format(window))
#     plt.ylabel('Reward')
#     plt.xlabel('Episode')

#     plt.tight_layout(pad=2)
#     plt.show()
    

gw.print_gw()
meanQ = np.mean(np.array(Q_list), axis = 0)
meanv = np.mean(meanQ, axis = 2)
#print(meanQ)
gw.print_v(meanv)
meanpi = np.argmax(meanQ, axis = 2)
gw.print_policy(meanpi)
E_reward, E_risk = estimate_rr(meanpi)
print "Mean reward ", E_reward
print "Risk probab ", E_risk
```
